**Preparing dataset for visualization**

Creating a new data table to facilitate ploting of time series chart.

```r

sales <- sales_data %>% select(Sales, Month, Hour) %>% group_by(Month, Hour) %>%
  summarise(Sales = sum(Sales)) %>%
  arrange(Month, Hour)

head(sales)
tail(sales)
```

**Sales Over the Year**

The charts below shows the frequency of monthly sales over the year.

```r
# line chart for sales over the year
ggplot(data = sales, aes(x = Month, y = Sales )) +
  geom_line() +
  labs(title = "Sales Over the Year", y= "Sales") + scale_x_continuous(breaks = seq(1, 12, by = 1)) +scale_y_continuous(labels = scales::comma_format(accuracy = 0.1))
```

```r
# Bar chart for sales over the year
ggplot(data = sales_data, aes(x = Month, fill = Sales, group = Sales, stat = "identity" )) + 
  geom_bar() +
  labs(title = "Sales over time", x= "Month", y= "Sales") + scale_x_continuous(breaks = seq(1, 12, by = 1))
```

According to the chart the month of December is when sales is most high.

**Most Sold Product**

The chart below shows the most sold products for a year.

```r
# bar chart for Most Sold Product
ggplot(data = sales_data, aes(x = Product, fill= Product )) +
  geom_bar() +
  labs(title = "Most Sold Product", y= "Sales") + theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

From the chart above, it can be seen that he product USB-C Charging Cable, Lightning Charging Cable, AAA Batteries (4-pack), AA Batteries (4-pack), Wired Headphones are the top 5 most sold product.

**Most City with Sales**

The Bar chart below shows the frequency of sales in each city over the period of a year.

```r
ggplot(data = sales_data, aes( y = City, fill= City )) +
  geom_bar() +
  labs(title = "Most City with Sales", x= "Sales") + theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

 The city with the most sales is San Francisco.

**Sales over a Day**

 The chart below shows the frequency of sales through out the day.

```r
ggplot(data = sales, aes(x = Hour, y = Sales, )) +
  geom_line() +
  labs(title = "Sales over a Day", y= "Sales") + scale_x_continuous(breaks =, seq(0, 23, , by = 1)) + scale_y_continuous(labels = scales::comma_format(accuracy = 0.1))
  
```

According to the chart, the time range in which sales most high is from 10am to 8pm.
