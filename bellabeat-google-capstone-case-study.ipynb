Here, i will be processing the data for analysis using R programming language. This will involve importing the datasets, cleaning the data and merging different dataset for analysis.
```r
# improting of dataset into R

 Activity <- read.csv("/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv")
Sleepday <- read.csv("/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv")
Weight_info <- read.csv("/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/weightLogInfo_merged.csv")
```

```r
# lets take a look at that each dataset

## preview of the first 6 row of each dataframe
head(Activity)
head(Sleepday)
head(Weight_info)

## lets check the data structure of each dataframe
str(Activity)
str(Sleepday)
str(Weight_info)

```

```r
# here i am going to check the sample size of each dataset

Activity %>%
  count(Id) %>%
  nrow()

Sleepday %>%
  count(Id) %>%
  nrow()

Weight_info %>%
  count(Id) %>%
  nrow()
```

As noted above, the sample size of the weight_info data is too small to be used for analysis. After some consideration, I have decided to drop the dataset as it could affect the overall sample size significantly when merging different datasets, potentially distorting my analysis.

**Cleaning of datasets**

This involves checking each dataset for any duplicate, missing values ormisspellings and making any correction where neccesary.

```r
# checking for duplicate

sum(duplicated(Activity))
sum(duplicated(Sleepday))
```

```r
# removing of duplicate from sleepday data

Sleepday.v2 <- Sleepday%>%
  distinct()

## Making sure duplicate is removed

sum(duplicated(Sleepday.v2))
```

```r
# checking for missing values

sum(is.na(Activity))
sum(is.na(Sleepday.v2))
```

**Merge datasets**

This involves preparing each dataset for merging by making sure the data point have the same and correct data structure.


```r
# preparing data for merging

## clean each dateset colunm name and change data type of each dataset from Character to Date formate

Activity.v2 <- Activity %>%
  rename(Date=ActivityDate) %>%
  rename_with(tolower) %>%
  clean_names() %>%
  mutate(date = as.Date(date, format = "%m/%d/%Y"))

Sleepday.v3 <- Sleepday.v2 %>% 
  rename(Date=SleepDay) %>%
  rename_with(tolower) %>%
  clean_names() %>%
  mutate(date = as.Date(date, format = "%m/%d/%Y"))  

## checking changes made

head(Activity.v2)
head(Sleepday.v3)

```

```r
# merging of datasets

User_usage <- merge(Activity.v2, Sleepday.v3, by= c("id", "date"))

# preview new dataset

head(User_usage)
```

# **Analyze**



```r
# Adding a new total active minutes column to know the daily active time of user

User_usage$totalactiveminutes <- User_usage$veryactiveminutes + User_usage$fairlyactiveminutes + User_usage$lightlyactiveminutes


# User_usage dataset summary

User_usage %>%
  select(totalsteps,totaldistance,veryactiveminutes,fairlyactiveminutes,
         lightlyactiveminutes,sedentaryminutes,calories,totalsleeprecords,
         totalminutesasleep,totaltimeinbed,totalactiveminutes) %>%
  summary()
```

**Observation for the summary of the dataset**

1. The average steps by users per day is 8,515.
2. The average total daily active minutes of users is 259.5 min(approximately 4 hours). 
3. The average daily sedentary minutes of users is 712.1 min(approximately 12 hours).
4. The average daily calories burnt by users is 2389.
5. An average of 419.2 min or 7 hours is spent asleep by users and an average of 458.5 or 7 hours 30 min is spent in bed.


# **Share**

This process involves sharing of data through visualizations, which will be showing the correlation between different data points in the dataset.

```r
# Correlation between the steps taken per day and calorie burned 

ggplot(data = User_usage, aes(x = totalsteps, y = calories)) +
  geom_point(color = 'Purple') +
  geom_smooth(method = "lm", se = FALSE)+
  labs(title = "Correlation Between Total steps and Daily Calories", x= "Steps", y= "Daily calories")

```

According to the chart above, there is a positve correlation between the amount of steps taken and calories burned.

```r
# Correlation between total active minutes per day and daily calories

ggplot(data = User_usage, aes(x = totalactiveminutes, y = calories)) +
  geom_point(color = 'Red') +
  geom_smooth(method = "lm", se = FALSE, color = 'Green',)+
  labs(title = "Correlation Between Total Active Minutes and Daily Calories", x= "Active Time(Min)", y= "Daily calories")
```

The above scatter plot shows a positive correlation between the active minutes per day and the calories burned.

```r
# Correlation between sedentary minutes per day and daily calories

ggplot(data = User_usage, aes(x = sedentaryminutes, y = calories)) +
  geom_point(color = 'purple') +
  geom_smooth(method = "lm", se = FALSE, color = 'yellow')+
labs(title = "Correlation Between Sedentary Minutes and Daily Calories", x= "Sedentary Time (min)", y= "Daily calories")
```

The above scatter plot shows a positive correlation between the sedentary time per day and the calories burned.

```r
# Correlation between total minutes asleep and total time in bed

ggplot(data = User_usage, aes(x = totalminutesasleep, y = totaltimeinbed)) +
  geom_point(color = 'Blue') +
  geom_smooth(method = "lm", se = FALSE, color = 'Orange')+
  labs(title = "Correlation Between Total Minutes Asleep and Total Time in Bed", x="Time asleep (min)", y="Total time in bed (min)")
```

The scatter plot displays a positive correlation between time asleep and time in bed.
